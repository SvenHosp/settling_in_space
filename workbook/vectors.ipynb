calc intersection point and proof!

```python
def get_tuples(p1, d1, p2, d2):
    t1 = None
    t2 = None
    for axis in range(0, d2.size):
        if d2[axis] != 0:
            t1 = axis
            break

    for axis in range(0, d1.size):
        if d1[axis] != 0 and axis != t1:
            t2 = axis
            break
    
    axis_tuple = (t1, t2) if t1 is not None and t2 is not None else None
    
    axis_out = None
    for axis in range(0, d1.size):
        if axis != t1 and axis != t2:
            axis_out = axis
            
    return axis_tuple, axis_out
    

def get_scalars(p1, d1, p2, d2, axis_tuple = (0, 1)):
    if axis_tuple is None:
        return None, None
    _I = axis_tuple[0]
    _II = axis_tuple[1]
    a_t = p2[_I] * d2[_II] - p2[_I] * d2[_I] - p1[_I] * d2[_II] + p1[_II] * d2[_I]
    b_t = d1[_I] * d2[_II] - d1[_II] * d2[_I]
    
    scalar_t = a_t/b_t
    
    axis_s = None
    
    for axis in range(0, len(axis_tuple)):
        if d2[axis] != 0:
            axis_s = axis
            break
    
    scalar_s = (p1[axis_s] + scalar_t * d1[axis_s] - p2[axis_s])/d2[axis_s] if axis_s is not None else None
    
    return scalar_t, scalar_s

def proof_skew(p1, d1, p2, d2, t, s, axis):
    if t is None or s is None:
        return False
    return p1[axis] + t * d1[axis] == p2[axis] + s * d2[axis]

def get_intersection_point(p1, d1, p2, d2):
    axis_tuple, proof_axis = get_tuples(p1, d1, p2, d2)
    
    t, s = get_scalars(p1, d1, p2, d2, axis_tuple = axis_tuple)

    is_intersection_point_possible = proof_skew(p1, d1, p2, d2, t, s, proof_axis) if axis_tuple is not None else False

    intersection_point = p1 + t * d1 if is_intersection_point_possible else None
    
    return is_intersection_point_possible, t, s, intersection_point
```

```python
intersection, t, s, i_p = get_intersection_point(p_1, d_1, p_2, d_2)
if intersection:
    print(i_p)
else:
    print('no point')
```

calculate interseption course

```python
def calc_intersect(ray_point, ray_dir, sphere_centre, sphere_radius):
    unit = numpy.ones(3)

    oc = ray_point - sphere_centre
    a = numpy.dot(ray_dir, ray_dir)
    b = 2 * numpy.dot(oc, ray_dir)
    c = numpy.dot(oc, oc) - sphere_radius*sphere_radius
    discrim = b*b - 4*a*c

    if discrim > 0:
        print('solution exists')
        square_root = numpy.sqrt(discrim)
        t0 = (-b - square_root) / (2*a)
        t1 = (-b + square_root) / (2*a)
        if t0 < t1:
            return t0
        else:
            return t1
    else:
        print('no solution exists')
```

```python
# https://www.forum-marinearchiv.de/smf/index.php?topic=27157.0
# https://www.codefull.net/2015/06/intersection-of-a-ray-and-a-line-segment-in-3d/
# https://gist.github.com/danieljfarrell/faf7c4cafd683db13cbc
# https://stackoverflow.com/questions/2316490/the-algorithm-to-find-the-point-of-intersection-of-two-3d-line-segment
# a (V1 X V2) = (P2 - P1) X V2
import numpy
e_pos = numpy.array([1,1,1])
e_dir = numpy.array([0,1,0])
e_speed = 1

s_pos = numpy.array([5,1,1])
s_speed = 2

e_s_pre = e_pos - s_pos
e_s = 1/numpy.linalg.norm(e_s_pre)*e_s_pre

e_pos1 = e_pos + e_speed*e_dir

dist = calc_intersect(e_pos1,e_s,s_pos,s_speed)

sphere_point = e_pos1 + dist*e_s

s_dir_pre = sphere_point - s_pos
s_dir = 1/numpy.linalg.norm(s_dir_pre) * s_dir_pre

intersection, t, s, i_p = get_intersection_point(e_pos, e_dir, s_pos, s_dir)
if intersection:
    print(i_p)
else:
    print('no point')
```
