```python
import numpy
# line e_pos + a * e_dir
p_1 = numpy.array([1,1,1])
d_1 = numpy.array([1,0,0])

# line s_pos + b * s_dir
p_2 = numpy.array([5,5,1])
d_2 = numpy.array([0,1,0])
```

proof collinearity

```python
r = d_1/d_2

if r[0] == r[1] and r[1] == r[2]:
    print('collinear')
else:
    print('not collinear')
```

calc intersection point and proof!

```python
def get_tuples(p1, d1, p2, d2):
    t1 = None
    t2 = None
    for axis in range(0, d2.size):
        if d2[axis] != 0:
            t1 = axis
            break

    for axis in range(0, d1.size):
        if d1[axis] != 0 and axis != t1:
            t2 = axis
            break
    
    axis_tuple = (t1, t2) if t1 is not None and t2 is not None else None
    
    axis_out = None
    for axis in range(0, d1.size):
        if axis != t1 and axis != t2:
            axis_out = axis
            
    return axis_tuple, axis_out
    

def get_scalars(p1, d1, p2, d2, axis_tuple = (0, 1)):
    if axis_tuple is None:
        return None, None
    _I = axis_tuple[0]
    _II = axis_tuple[1]
    a_t = p2[_I] * d2[_II] - p2[_I] * d2[_I] - p1[_I] * d2[_II] + p1[_II] * d2[_I]
    b_t = d1[_I] * d2[_II] - d1[_II] * d2[_I]
    
    scalar_t = a_t/b_t
    
    for axis in range(0, len(axis_tuple)):
        if d2[axis] != 0:
            axis_s = axis
            break
    
    scalar_s = (p1[axis_s] + scalar_t * d1[axis_s] - p2[axis_s])/d2[axis_s]
    
    return scalar_t, scalar_s

def proof_skew(p1, d1, p2, d2, t, s, axis):
    return p1[axis] + t * d1[axis] == p2[axis] + s * d2[axis]

def get_intersection_point(p1, d1, p2, d2):
    axis_tuple, proof_axis = get_tuples(p1, d1, p2, d2)
    
    t, s = get_scalars(p1, d1, p2, d2, axis_tuple = axis_tuple)

    is_intersection_point_possible = proof_skew(p1, d1, p2, d2, t, s, proof_axis) if axis_tuple is not None else False

    intersection_point = p1 + t * d1 if is_intersection_point_possible else None
    
    return is_intersection_point_possible, t, s, intersection_point
```

```python
intersection, t, s, i_p = get_intersection_point(p_1, d_1, p_2, d_2)
if intersection:
    print(i_p)
else:
    print('no point')
```
