```python
# http://viclw17.github.io/2018/07/16/raytracing-ray-sphere-intersection/
bool hit_sphere(const vec3& center, float radius, const ray& r){
    vec3 oc = r.origin() - center;
    float a = dot(r.direction(), r.direction());
    float b = 2.0 * dot(oc, r.direction());
    float c = dot(oc,oc) - radius*radius;
    float discriminant = b*b - 4*a*c;
    return (discriminant>0);
}

float hit_sphere(const vec3& center, float radius, const ray& r){
    vec3 oc = r.origin() - center;
    float a = dot(r.direction(), r.direction());
    float b = 2.0 * dot(oc, r.direction());
    float c = dot(oc,oc) - radius*radius;
    float discriminant = b*b - 4*a*c;
    if(discriminant < 0){
        return -1.0;
    }
    else{
        return (-b - sqrt(discriminant)) / (2.0*a);
    }
}
```

```python
def calc_intersect(ray_point, ray_dir, sphere_centre, sphere_radius):
    print('{0} {1} {2} {3}'.format(ray_point, ray_dir, sphere_centre, sphere_radius))
    unit = numpy.ones(3)

    oc = ray_point - sphere_centre
    a = numpy.dot(ray_dir, ray_dir)
    b = 2 * numpy.dot(oc, ray_dir)
    c = numpy.dot(oc, oc) - sphere_radius*sphere_radius
    discrim = b*b - 4*a*c

    if discrim > 0:
        print('solution exists')
        square_root = numpy.sqrt(discrim)
        t0 = (-b - square_root) / (2*a)
        t1 = (-b + square_root) / (2*a)
        if t0 < t1:
            return t0
        else:
            return t1
    else:
        print('no solution exists')
```
