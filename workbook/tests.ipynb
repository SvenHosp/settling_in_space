```python
import numpy as geek 
import matplotlib.pyplot as plot 
   
# 1D Array as per Gaussian Distribution 
mean = 0 
std = 0.1
array = geek.random.normal(0, 0.1, 100) 
```

```python
# Source Code :  
# https://docs.scipy.org/doc/numpy-1.13.0/reference/ 
# generated/numpy-random-normal-1.py 
count, bins, ignored = plot.hist(array, 60) 
plot.plot(bins, 1/(std * geek.sqrt(2 * geek.pi)) *
          geek.exp( - (bins - mean)**2 / (2 * std**2) ), 
          linewidth=2, color='r') 
plot.show() 
```

```python
for i in range(-50,50):
    j = i/100
    print(1/(std * geek.sqrt(2 * geek.pi)) *
          geek.exp( - (j - mean)**2 / (2 * std**2) ))
```

```python
import numpy as geek 
import matplotlib.pyplot as plot 
   
# 1D Array as per Gaussian Distribution 
mean = 6
std = 10
array = geek.random.normal(6, 3.0, 40)
for i in range(0,40):
    value = int(array[i])
    if value < 3 or value > 12:
        value = 6
    print(value)
    array[i] = value
# Source Code :  
# https://docs.scipy.org/doc/numpy-1.13.0/reference/ 
# generated/numpy-random-normal-1.py 
count, bins, ignored = plot.hist(array, 60)
plot.plot(bins, 1/(std * geek.sqrt(2 * geek.pi)) *
          geek.exp( - (bins - mean)**2 / (2 * std**2) ), 
          linewidth=2, color='r') 
plot.show() 
```
